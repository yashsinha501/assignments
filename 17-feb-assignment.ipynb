```python
# ANS 1 =>

# MongoDB is a popular document-oriented NoSQL database system that stores data in JSON-like documents with dynamic schemas.
# It is designed to scale out and handle large amounts of data, with features like sharding and replication built-in.

# Non-relational databases, also known as NoSQL databases, are databases that don't rely on the traditional relational model of 
# organizing data into tables with predefined schemas. Instead, they use a variety of data models, including document-oriented,
# key-value, graph, and column-family stores, among others.

# Non-relational databases are often preferred in scenarios where the data is unstructured or semi-structured, and where the 
# flexibility of schema-less data models can simplify development and data management. They are also useful in cases where the
# data is distributed across multiple servers, and where horizontal scaling is necessary.

# MongoDB is often preferred over SQL databases in scenarios where:

# There is a need to handle large amounts of unstructured or semi-structured data, as it can store data in flexible and dynamic
# schemas without sacrificing performance.

# There is a need for horizontal scaling, as MongoDB's sharding and replication features make it easy to scale out to handle large 
# amounts of data.

# There is a need for real-time data processing and analysis, as MongoDB's native support for indexing and aggregation makes it easy 
# to perform complex queries and analytics on large datasets.

# There is a need for a flexible data model that can evolve over time, as MongoDB's dynamic schemas make it easy to add or modify 
# fields as the application changes.

```

```python
# ANS 2 =>

# MongoDB is a popular NoSQL database system with a number of features that make it a popular choice for developers working on modern,
# data-driven applications. Some of the key features of MongoDB include:

# Document-oriented data model: MongoDB stores data in flexible, JSON-like documents that can have nested structures, arrays, and other
# data types. This allows developers to easily store and query complex data structures without having to map them to a fixed table 
# schema.

# Dynamic schema: MongoDB's documents have dynamic schemas, meaning that each document can have its own unique set of fields and data
# types. This allows for more flexibility in data modeling and makes it easy to add or remove fields as the application evolves.

# Indexing and aggregation: MongoDB supports flexible indexing and aggregation features that make it easy to query large datasets and 
# perform complex analytics. This includes support for geospatial queries, text search, and the ability to perform aggregations on the
# server side.

# Replication and sharding: MongoDB provides built-in support for replication and sharding, allowing developers to easily scale their
# databases horizontally across multiple servers. This provides high availability and fault tolerance, as well as improved performance.

# Native drivers and language support: MongoDB provides native drivers and language support for a variety of programming languages,
# including Java, Python, Node.js, and Ruby. This makes it easy for developers to integrate MongoDB with their preferred development
# stack.

# GridFS: MongoDB includes a file storage system called GridFS, which allows developers to store and retrieve large files, such as 
# images and videos, directly in the database. This eliminates the need for a separate file storage solution and makes it easier to 
# manage and scale file storage.


```

```python
# ANS 7 =>

# In MongoDB, delete_one(), delete_many(), and drop() are used to remove documents or collections from the database.

# delete_one() is used to remove a single document that matches a specified filter. It takes a filter document as a parameter,
# which specifies the criteria for the documents to be deleted. If there are multiple documents that match the filter, delete_one()
# will only delete the first matching document. If no document matches the filter, delete_one() does nothing.

# delete_many() is similar to delete_one(), but it deletes all documents that match the specified filter. It takes a filter document
# as a parameter, which specifies the criteria for the documents to be deleted. If no document matches the filter, delete_many() does
# nothing.

# drop() is used to remove an entire collection from the database. It takes no parameters and simply deletes the entire collection,
#including all the documents it contains.This method is irreversible,and the data cannot be recovered once the collection is dropped.

# These methods are useful for maintaining data consistency and cleaning up the database. delete_one() and delete_many() can be used 
# to selectively remove documents that are no longer needed,while drop() can be used to remove an entire collection that is no longer
# needed. It is important to use these methods with caution, as the data that is deleted cannot be recovered.
```
