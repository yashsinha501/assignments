```python
# ANS 2 =>

# In Python, the close() function is used to close a file object. 
# It is important to close a file after it has been used because it frees up system resources that were being used 
# to keep the file open.

# The close() function ensures that all resources associated with the file are released,including any buffers that may be
# in use. This ensures that the data in the file is saved properly and that the file is closed in a clean and safe manner.

# In addition to the resource management benefits, using the close() function is also a good programming practice because it 
# makes your code easier to understand and maintain. When you close a file explicitly, it makes it clear that you are done
# with the file and that other parts of the program should not attempt to access it.

# In summary, it is important to use the close() function in Python to properly manage system resources and ensure that 
# files are closed in a clean and safe manner.


```

```python
# ANS 5 =>

# In Python, the with statement is used with open() to create a context in which the file can be safely opened and closed 
# automatically. The with statement creates a block of code where the file is opened, used, and then automatically closed,
# without the need for an explicit call to the close() method. This ensures that the file is properly closed and resources are
# freed even if an error occurs in the code.

# Here is an example of using the with statement with open() to read a file:

# Example => 

# with open('example.txt', 'r') as file:
#     data = file.read()
#     print(data)

# In this example, the with statement creates a context in which the file object is created and opened using the open() method.
# The as keyword assigns the file object to a variable that can be used within the block of code. Once the block of code is
# exited, the file object is automatically closed and resources are freed, even if an exception occurs.

# The advantage of using the with statement with open() is that it ensures that the file is closed properly and resources are 
# freed automatically, without the need for an explicit call to the close() method. This makes the code simpler, safer, and 
# easier to read and understand. Additionally, it helps to avoid common errors such as forgetting to close the file, or
# accidentally leaving the file open for too long.
```
