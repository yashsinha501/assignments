```python
# ANS 2 =>

from flask import Flask
from flask import request

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello World!!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
    


```

```python
# ANS 3 => 

# In Flask, "routing" refers to the process of matching a URL to a specific function in your application. 
# An "app route" is a decorator in Flask that is used to bind a URL to a function.

# Here's an example of an app route in Flask:

# from flask import Flask

# app = Flask(__name__)

# @app.route('/')
# def index():
#     return 'Hello, World!'


# In this example, the @app.route('/') decorator tells Flask that the index() function should be called when the root URL (/) 
# is requested. The function then returns the string "Hello, World!" as a response.

# We use app routes in Flask for several reasons:

# To make our application more organized and maintainable. By mapping URLs to specific functions, we can easily understand how
# our application is structured and modify it as necessary.

# To provide a user-friendly interface. App routes allow us to create URLs that are easy to remember and understand, making it
# more user-friendly.

# To build RESTful APIs. App routes allow us to build APIs that can be accessed by other applications and services over the 
# internet. By mapping URLs to specific functions, we can provide a clear and consistent API interface.
```
