```python
# Ans 3 => 

val=int(input(" Units of electricity consumed in a month : "))
amount=0;
l=[4.5,6,10,20]
i=0
while val>0:
    if val>100:
        amount=amount+(l[i]*100)
        val=val-100
    else:
        amount=amount+(l[i]*val)
        val=0
    i=i+1
print("Electricity bill of the month is ", amount)

```
