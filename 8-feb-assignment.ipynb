```python
# Ans 3 =>

# ABC stand for Abstract Base Class.The abc  module in Python is a part of the standard library and provides support
# for defining abstract base classes. 
# Abstract base classes are used as a blueprint for creating objects in a subclass, and are used to define a common i
# nterface for a set of related objects.
# An ABC is an abstract class that cannot be instantiated. Instead, it defines a set of methods that must be implemented by 
# any concrete class that inherits from the ABC. By using ABCs, developers can ensure that certain methods are present in all 
# subclasses, leading to a more consistent and predictable interface across the codebase.

# The abc module is used to define and register abstract base classes, and to check if a class is a subclass of an abstract
# base class.
```
